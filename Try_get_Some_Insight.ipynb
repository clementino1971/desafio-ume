# Importando Bibliotecas

```
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

# Getting Data

```
path_csv = 'sample15k_synth.csv'
data = pd.read_csv(path_csv)
features = data.keys()
features
```

# Separando os dados de Treino e Teste

```
# Separando dados por data
train,test = data.query("date < '2015-09-01'") , data.query("date >= '2015-09-01'")

# Dividindo features em categorias para possiveis analises
demo_features = list(features[1:7])
job_features = list(features[7:11])
income_features = list(features[11:13])
debt_features = list(features[13:16])

features_base = demo_features + job_features + income_features + debt_features

X,y= train[features_base],train.target
X_test,y_test = test[features_base],test.target


print('{} Samples for train \n{} Samples for test'
          .format(len(X),len(X_test)))
```
